Firstly i fetch all prices for all tokens 

```python
def fetch_historical_price(start_date,end_date,token_id):
    daterange = pd.date_range(start_date, end_date)
    dates=[]
    normal_dates=[]
    for single_date in daterange:
        date=(single_date.strftime("%d-%m-%Y"))
        dates.append(date)
    for single_date in daterange:
        date=(single_date.strftime("%Y-%m-%d"))
        normal_dates.append(date)
    token_prices=[]
    for x in range(len(dates)):
        try:
            row={}
            request = requests.get('https://api.coingecko.com/api/v3/coins/'+ token_id +'/history?date=' + dates[x]) 
            if request.status_code == 200:
                price=request.json()['market_data']['current_price']['usd']
                row['blocktime'] = normal_dates[x]
                row['Price']= price
                token_prices.append(row)
            elif request.status_code == 429:
                time.sleep(60)
                price=request.json()['market_data']['current_price']['usd']
                row['blocktime'] = normal_dates[x]
                row['Price']= price
                token_prices.append(row)
            
        except KeyError:
            print(f'Api exceed on {dates[x]}')
            print(request.json())
            
        
    return(token_prices)
    
```

```python
coins={'lido-staked-sol':'stSOL','lido-dao':'LDO','wrapped-steth':'wstETH','solana':'SOL','usd-coin':'USDC','tether':'USDT'}
```

```python
for x in coins.keys():    
    result=fetch_historical_price('2022-01-14','2022-01-18',x)
    result_df=pd.DataFrame(result)
    date = datetime.datetime.today().strftime('%Y-%m-%d')
    path="Prices\\" + coins[x]+"_prices.csv"
    result_df.to_csv(path, sep=',',header=False,mode='a', encoding='utf-8', index=False)
```

```python
token1_prices=pd.read_csv(r'Prices\stSOL_prices.csv')
token1_prices['blocktime']=pd.to_datetime(token1_prices['blocktime'],infer_datetime_format=True)
token1_prices['token']='stSOL'
token2_prices=pd.read_csv(r'Prices\SOL_prices.csv')#Change here
token2_prices['blocktime']=pd.to_datetime(token2_prices['blocktime'],infer_datetime_format=False)
token2_prices['token']='SOL'#Change here
pool_dates=pool['blocktime']
token1_prices=pd.merge_asof(pool_dates,token1_prices,on=['blocktime'])
token2_prices=pd.merge_asof(pool_dates,token2_prices,on=['blocktime'])
frames = [token1_prices,token2_prices]
prices = pd.concat(frames)
```

```python
TVL=pool.merge(prices,on=['blocktime','token'])
TVL=TVL.drop_duplicates(ignore_index=True)
TVL['usd_price']=TVL['balance']*TVL['Price']
TVL.to_csv(r'Google DS\New_tvl\Saber_SOL_pool.csv',sep=',',header=False,mode='a', encoding='utf-8',index=False)#Change here
```
